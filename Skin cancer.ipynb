## 5.2 Mathematical Analysis (if applicable)

This section provides a mathematical foundation for the deep learning approach used in this project. The methodology involves key deep learning concepts such as **Convolutional Neural Networks (CNNs), loss functions, backpropagation, and evaluation metrics**.

### **1. Convolutional Neural Networks (CNNs)**
A CNN applies convolution operations to extract hierarchical features from images. The **convolution operation** between an input image \( X \) and a filter (kernel) \( W \) can be mathematically represented as:

\[
Y(i, j) = \sum_m \sum_n X(i+m, j+n) \cdot W(m, n)
\]

where:
- \( X(i, j) \) is the pixel value at position \( (i, j) \),
- \( W(m, n) \) is the kernel applied over the input,
- \( Y(i, j) \) is the output feature map at position \( (i, j) \).

The convolutional layer captures spatial hierarchies and patterns such as edges, textures, and more complex structures as layers deepen.

### **2. Loss Function: Categorical Cross-Entropy**
The model is trained using the **categorical cross-entropy loss function**, commonly used for multi-class classification tasks. It is defined as:

\[
L = -\sum_{i=1}^{N} y_i \log(\hat{y}_i)
\]

where:
- \( N \) is the number of classes,
- \( y_i \) is the true label (one-hot encoded),
- \( \hat{y}_i \) is the predicted probability for class \( i \).

This loss function measures the difference between the true and predicted probability distributions, pushing the model to maximize the probability of the correct class.

### **3. Backpropagation and Optimization**
During training, gradients are computed using **backpropagation**, and weights are updated iteratively using an optimizer such as **Adam**. The update rule for weight \( w_t \) at iteration \( t \) using Adam optimizer is:

\[
m_t = \beta_1 m_{t-1} + (1 - \beta_1) \nabla L
\]

\[
v_t = \beta_2 v_{t-1} + (1 - \beta_2) (\nabla L)^2
\]

\[
\hat{m}_t = \frac{m_t}{1 - \beta_1^t}, \quad \hat{v}_t = \frac{v_t}{1 - \beta_2^t}
\]

\[
w_{t+1} = w_t - \frac{\alpha \hat{m}_t}{\sqrt{\hat{v}_t} + \epsilon}
\]

where:
- \( m_t \) and \( v_t \) are the first and second moment estimates,
- \( \beta_1 \) and \( \beta_2 \) are decay rates,
- \( \alpha \) is the learning rate,
- \( \epsilon \) is a small constant for numerical stability.

Adam dynamically adjusts the learning rate for each weight, ensuring faster convergence and reducing oscillations.

### **4. Evaluation Metrics**
The performance of the model is assessed using standard classification metrics:

#### **Accuracy**
\[
\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
\]
where:
- \( TP \) = True Positives, \( TN \) = True Negatives,
- \( FP \) = False Positives, \( FN \) = False Negatives.

#### **Precision**
\[
\text{Precision} = \frac{TP}{TP + FP}
\]
Precision measures the proportion of correctly predicted positive cases among all predicted positives.

#### **Recall (Sensitivity)**
\[
\text{Recall} = \frac{TP}{TP + FN}
\]
Recall indicates how well the model identifies actual positive cases.

#### **F1-Score**
\[
\text{F1-Score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}
\]
The F1-score balances precision and recall, ensuring that both false positives and false negatives are accounted for.

#### **ROC-AUC Score**
\[
\text{AUC} = \int_{0}^{1} TPR(FPR) \, dFPR
\]
where:
- \( TPR \) (True Positive Rate) and \( FPR \) (False Positive Rate) are plotted to evaluate model performance across different classification thresholds.

These metrics ensure a **comprehensive evaluation** of the model’s effectiveness in skin cancer classification.

### **Conclusion**
By leveraging CNNs for feature extraction, optimizing model parameters using the Adam optimizer, and evaluating performance with classification metrics, this project aims to build an **accurate, interpretable, and clinically relevant deep learning model for automated skin cancer detection**

**Experiment 1: Baseline Model (Flawed) - Explanation**

1.	Image Resizing (128x128)
In this experiment, all images are resized to 128x128 pixels to standardize the input dimensions for the deep learning model. Resizing helps reduce computational costs while maintaining essential visual features necessary for classification.

2.	Data Leakage: A Critical Flaw
One major issue with this baseline model is data leakage. In this case, shuffling and data augmentation were performed before splitting the dataset into training and validation sets.


•	Why is this bad? Data augmentation (e.g., flipping, rotating, scaling images) is used to artificially expand the training set. However, if augmentation is applied before splitting, transformed versions of the same images may end up in both the training and validation sets.

•	Impact: The model is indirectly exposed to information from the validation set during training, leading to overestimated accuracy that does not generalize well to unseen data.

4.	Dropout (0.3)
Dropout is a regularization technique used to prevent overfitting by randomly disabling neurons during training.
•	In this baseline experiment, a Dropout rate of 0.3 means 30% of neurons in certain layers are randomly deactivated during training, making the model less dependent on specific neurons and improving generalization.
5. Limited Epochs (50)
The model is trained for only 50 epochs, meaning it gets a limited opportunity to learn patterns from the data.
•	Fewer epochs generally result in underfitting, where the model fails to learn enough meaningful representations from the data.
6. Key Issue: Unreliable Results
Due to data leakage, the model may report artificially high accuracy but fail on real unseen data.
•	Corrective Action:
o	Ensure proper train-validation split first, then apply augmentation only on the training data in subsequent experiments.
o	Introduce better dropout rates and fine-tuned learning rates to improve generalization.
This flawed baseline serves as a benchmark to demonstrate how proper preprocessing and training techniques significantly improve model performance in later experiments.

```python
from sklearn.utils import shuffle
from tensorflow.keras.preprocessing.image import ImageDataGenerator

X_train, y_train = load_data('/kaggle/input/skin-cancer-medical-images/archive/train/train','/kaggle/input/skin-cancer-medical-images/archive/train_labels.csv')
X_test = load_test_data('/kaggle/input/skin-cancer-medical-images/archive/test/test')

#shuffling the data
X_train, y_train = shuffle(X_train, y_train)


#data Augmentation
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
datagen.fit(X_train) 

print(X_train.shape, y_train.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

# Encode labels using LabelEncoder
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)


# Convert labels to one-hot encoding
num_classes = len(label_encoder.classes_)
y_train_one_hot = to_categorical(y_train_encoded, num_classes)


# Print the shapes of the data
print(X_train.shape, y_train_one_hot.shape)
```

```python
#splitting the data into training and validation
X_train_new, X_val, y_train_new, y_val = train_test_split(
    X_train, y_train_one_hot, test_size=0.2, random_state=42, stratify=y_train 
)

print("X_train_new shape:", X_train_new.shape)
print("y_train_new shape:", y_train_new.shape)
print("X_val shape:", X_val.shape)
print("y_val shape:", y_val.shape)
```

```python
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.applications import EfficientNetB3  # Or V2B2
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, BatchNormalization # Import layers
from tensorflow.keras.models import Model
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.utils import class_weight
from tensorflow.keras import backend as K
import numpy as np




K.clear_session()

base_model = EfficientNetB3(weights='imagenet', include_top=False, input_shape=(image_size , image_size , 3)) # Or V2B2

x = GlobalAveragePooling2D()(base_model.output)
x = Dense(256, activation='relu', kernel_regularizer=l2(0.001))(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)
x = Dense(num_classes, activation='softmax')(x)  # Output layer

model = Model(inputs=base_model.input, outputs=x)

for layer in base_model.layers[80:]: # Example: Unfreeze after layer 100
    layer.trainable = True

model.compile(optimizer=Adam(learning_rate=0.0001), # Lower learning rate
              loss='categorical_crossentropy', # Or sparse_categorical_crossentropy
              metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience = 50, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=3, min_lr=0.00001)

history = model.fit(
    datagen.flow(X_train_new, y_train_new, batch_size=32),  # Adjust batch size
    epochs=50,  # Or more
    validation_data=(X_val, y_val),
    callbacks=[early_stopping, reduce_lr],
    steps_per_epoch=len(X_train) // 32 # Important to avoid errors.
)


# Evaluate the model on the test set
loss, accuracy = model.evaluate(X_val, y_val)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```

```python
import os
import os.path as osp
import numpy as np
from PIL import Image
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
image_size = 224
#image_size =384
def load_data(image_dir,labels_csv,img_size=(image_size , image_size)):
    images = []
    labels = []
    for filename in os.listdir(image_dir):
        if filename.endswith('.jpg'):  # Specifically check for .jpg
            image_id = filename[:-4]  # Remove ".jpg"
            try:
                label = df[df['image_id'] == image_id]['label'].values[0]
                image_path = osp.join(image_dir, filename)
                img = Image.open(image_path)
                img = img.resize(img_size)
                img = img.convert('RGB')
                img = np.asarray(img, dtype=np.float32) / 255.0
                images.append(img)
                labels.append(label)
            except FileNotFoundError:
                print(f"Image not found: {image_path}")
            except IndexError:
                print(f"Label not found for image: {image_id}")
            except Exception as e:
                print(f"Error processing image {image_path}: {e}")
    X = np.stack(images)
    y = np.array(labels)
    return X, y

def load_test_data(image_dir,img_size=(image_size ,image_size)):
    images = []
    for filename in os.listdir(image_dir):
        if filename.endswith('.jpg'):  # Specifically check for .jpg
            image_id = filename[:-4]  # Remove ".jpg"
            try:
                image_path = osp.join(image_dir, filename)
                img = Image.open(image_path)
                img = img.resize(img_size)
                img = img.convert('RGB')
                img = np.asarray(img, dtype=np.float32) / 255.0
                images.append(img)
            except FileNotFoundError:
                print(f"Image not found: {image_path}")
            except Exception as e:
                print(f"Error processing image {image_path}: {e}")

    X = np.stack(images)
    return X
```

```python
test_dir = '/kaggle/input/skin-cancer-medical-images/archive/test/test'

image_ids = []  # Initialize an empty list to store the IDs

for filename in os.listdir(test_dir):
    if filename.endswith('.jpg'):
        image_id = filename[:-4]
        image_ids.append(image_id)

# Create a Pandas DataFrame from the list of IDs
submission_df = pd.DataFrame({'id': image_ids})
```

```python
from sklearn.utils import shuffle
from tensorflow.keras.preprocessing.image import ImageDataGenerator

X_train, y_train = load_data('/kaggle/input/skin-cancer-medical-images/archive/train/train','/kaggle/input/skin-cancer-medical-images/archive/train_labels.csv')
X_test = load_test_data('/kaggle/input/skin-cancer-medical-images/archive/test/test')



print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,15))
for i in range(16):
    plt.subplot(4,4, i+1)
    plt.axis('off')
    plt.imshow(X_train[i])
    #the title should display the label
    plt.title(y_train[i])
```

```python
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

# Encode labels using LabelEncoder
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)


# Convert labels to one-hot encoding
num_classes = len(label_encoder.classes_)
y_train_one_hot = to_categorical(y_train_encoded, num_classes)


# Print the shapes of the data
print(X_train.shape, y_train_one_hot.shape)
```

```python
from sklearn.utils import class_weight


# Calculate class weights
class_weights = class_weight.compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weight_dict = dict(zip(np.unique(y_train), class_weights))
```

```python
#splitting the data into training and validation
X_train_new, X_val, y_train_new, y_val = train_test_split(
    X_train, y_train_one_hot, test_size=0.2, random_state=42, stratify=y_train 
)

print("X_train_new shape:", X_train_new.shape)
print("y_train_new shape:", y_train_new.shape)
print("X_val shape:", X_val.shape)
print("y_val shape:", y_val.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
#shuffling the data
X_train_new, y_train_new = shuffle(X_train_new, y_train_new)

#data Augmentation

datagen = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

datagen.fit(X_train_new) 

print(X_train_new.shape, y_train_new.shape)
```

```python
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.applications import EfficientNetB3  # Or V2B2
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, BatchNormalization # Import layers
from tensorflow.keras.models import Model
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.utils import class_weight
from tensorflow.keras import backend as K
import numpy as np

K.clear_session()
base_model = EfficientNetB3(weights='imagenet', include_top=False, input_shape=(image_size , image_size , 3)) # Or V2B2


x = GlobalAveragePooling2D()(base_model.output)
x = Dense(256, activation='relu', kernel_regularizer=l2(0.01))(x)
x = BatchNormalization()(x)
x = Dropout(0.7)(x)
x = Dense(num_classes, activation='softmax')(x)  # Output layer

model = Model(inputs=base_model.input, outputs=x)

for layer in base_model.layers[80:]: # Example: Unfreeze after layer 100
    layer.trainable = True


model.compile(optimizer=Adam(learning_rate=0.0001), # Lower learning rate
              loss='categorical_crossentropy', # Or sparse_categorical_crossentropy
             metrics= ['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience = 50, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=3, min_lr=0.00001)

history = model.fit(
    datagen.flow(X_train_new, y_train_new, batch_size=32),  # Adjust batch size
    epochs=100,  # Or more
    validation_data=(X_val, y_val),
    callbacks=[early_stopping, reduce_lr],
    steps_per_epoch=len(X_train) // 32, # Important to avoid errors.
    class_weight=class_weight_dict  # Apply class weights

)
    


# Evaluate the model on the validation set
loss, accuracy = model.evaluate(X_val, y_val)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")

# Plot training and validation loss curves
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Validation'], loc='upper right')
plt.show()
```

```python
from sklearn.metrics import classification_report

# Predict probabilities for the test set
y_pred = np.argmax(model.predict(X_test),axis = 1)
pred_labels = label_encoder.inverse_transform(y_pred)
submission_df['target'] = pred_labels
submission_df.to_csv("submission.csv", index=False) 
```

## 6.3 Summary of improvements and alternative approaches

Throughout this project, multiple techniques were applied to optimize the performance of the deep learning model for skin cancer classification. Below is a summary of the improvements and alternative approaches considered:

* **Transfer Learning with EfficientNetB3:** Instead of training from scratch, a pre-trained EfficientNetB3 model was used, allowing the model to leverage pre-learned features, improving convergence speed and accuracy.
* **Fine-Tuning Strategy:** Initially, the base layers of EfficientNetB3 were frozen, and only the newly added layers were trained. Later, selective unfreezing of deeper layers (after layer 80) was performed to refine the model further.
* **Data Augmentation:** Applied transformations such as rotations, width and height shifts, shear, zoom, and horizontal flips to increase the dataset’s diversity and prevent overfitting. In Experiment 1, augmentation was applied before the train/validation split, leading to data leakage. This was corrected in Experiment 2 by ensuring augmentation occurred only after splitting.
* **Regularization Techniques:** Included Batch Normalization and L2 Regularization to stabilize training and reduce overfitting risks.
* **Hyperparameter Tuning:** Various parameters such as dropout rates, batch sizes, and learning rates were adjusted to improve model performance.
* **Class Imbalance Handling:** The dataset exhibited an imbalance in certain classes, which was managed through stratified sampling during data splitting and appropriate augmentation strategies.
* **Early Stopping & Learning Rate Reduction:** Monitored validation loss to stop training when improvements plateaued while adjusting the learning rate dynamically to refine optimization.
* **Alternative Models Considered:** While EfficientNetB3 was chosen due to its strong performance in medical image classification, models like InceptionV3 were also evaluated for potential use.

Despite these optimizations, some challenges remain, such as ensuring generalizability across diverse patient populations and improving model interpretability. Future work may explore ensemble learning methods or Generative Adversarial Networks (GANs) for synthetic data augmentation.
